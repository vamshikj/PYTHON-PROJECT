```python
# 7.On the Discrete Variable  Find the relationship between Discrete and Sales price

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data from the Excel file
data = pd.read_excel("/content/HousePrediction.xlsx")

# Select the discrete variable and sales price columns
MSSubClass = data["MSSubClass"]
sales_price = data["SalePrice"]

# Calculate the correlation coefficient
correlation = MSSubClass.corr(sales_price)

# Print the correlation coefficient
print(f"Correlation between discrete variable and sales price: {correlation}")

# Create a scatter plot
plt.scatter(MSSubClass, sales_price)
# Add labels and title
plt.xlabel("Discrete Variable (MSSubClass)")
plt.ylabel("Sales Price")
plt.title("Scatter Plot of Discrete Variable vs. Sales Price")

# Show the plot
plt.show()
```

```python


import pandas as pd
from scipy import stats

# Read the data from the Excel file
data = pd.read_excel("/content/HousePrediction.xlsx")

# Check the data type of the data variable
if data.dtypes.apply(lambda x: x == str).any():
    # Convert the data variable to a DataFrame containing numeric data
    data = data.apply(pd.to_numeric)

# Calculate the z-score for each data point
z_scores = stats.zscore(data)

# Define a threshold for outlier identification
threshold = 3

# Identify outliers
outliers = data[abs(z_scores) > threshold]

# Print the outliers
print(outliers)
```
